## ***COLUMNS EXPLAINED*** ##

#### ***Descriptions*** ####
- ***Year  :*** *The year that the sepcific season took place in*                       
- ***Tm    :*** *The team that the player was signed to in that year*
- ***Age   :*** *Age of the player*
- ***Games :*** *Games played each season. (Max is 82)*
- ***MP    :*** *Minutes Played each season*
- ***PER   :*** *Player Efficiency Rating, This is a value that is calculated according to a player's per-minute productivity*

#### ***Shooting Attributes*** ####
- ***TS%   :*** *True Shooting Percentage, a percentage of a player's shooting efficiency which takes into account 2-pointer field goals, 3 pointers, and free throws*
- ***3PAr  :*** *3-Point Attempt Rate, a percentage of field goals attempts from beyond the 3 point line*
- ***FTr   :*** *This is not the regualr free throw percentage that we see in game, this is the number of FT attempts per FG attempt*

#### ***Other Attributes*** ####
- ***ORB%  :*** *Offensive Rebound Percentage, a percentage of offensive rebounds a player grabbed while he was on court*
- ***DRB%  :*** *Defensive Rebound Percentage, a percentage of defensive rebounds a player grabbed while he was on court*
- ***TRB%  :*** *Total Rebound Percentage*
- ***AST%  :*** *A percentage of assists achieved while on the court. An assist is attributed to a player when they pass to a teammate in a way that leads to a score by field goal*
- ***STL%  :*** *A pecentage of steals achieved by a single player while on the court*
- ***BLK%  :*** *A percentage of blocks achieved by a player where he blocks the field goal of the opponent while the ball is going up and before hitting the back board*
- ***TOV%  :*** *This porcentage is also different from the regular % shown in game, this is a percentage of turnovers commited per 100 plays*

#### ***Question 1:***  *Compared to the time in the NBA where Centers were dominating posting up with Post hooks and Sky hooks, is there evidence that shows how player are more revolved around shooting 3 pointers more?*


```python
graph = (sns.boxplot(x="Year", y='TS%',data=df))
#average
```

*This is a barplot representing the true shooting percentages of players throught out the 2011 - 2016 season. By the looks of it, we are able to recognize a small decrease as time goes towards 2016. The decrease in TS% can very much likely be caused by the increase in jumpshots taken by players. As the true shooting percentage is calculated by taking into count field goals as well as three pointer and freethrows, it is understandable that the decrease in TS% is due to players acclimating to taking more jumpshots from beyond the three point line or mid range shots.*

```python
sns.violinplot(data = df, x='Year', y='3PAr')
```

*This plot makes it somewhat more clear to prove that the NBA is slowly becoming a shooter's game .As time went on, the rate of attempted 3 pointers increased. We might even be able to predict that in a decade or so, almost all teams will be able to run 5 out, where all 5 players on the court can comfortably shoot the 3 ball, which will allow them to stand outside the 3 point line and spread out the floor*

#### ***Question 2:*** *Can the attributes of this dataset show the change in physicality of the NBA over the years?*

```python
reb = (sns.barplot(x="Year", y='FTr',data=df))
reb.set_ylabel("Free throw attempts")
```

*The barplot above is one way of answering this question with the FTr attribute within this dataset. As a reminder, FTr, which stands for Free Throw Rate, this is the number of free throw attempts per each field goal attempt. The change of physicallity in the NBA from vintage to modern times has always been a sizable topic. Some fans might argue that a dominant player from the current era would not be able to compete in the old days as the amount of contact back then was much higher than the amount of contact in the modern day NBA. As the attribute FTr shows the number of freethrows attempter per field goal attempt, we are able to see where the physicallity differ as more physicality results in more fouls, which results in players attempting more and more freethrows.*

*According to the Bar plot above, which shows the free throw rate, we are able to see that there is no gradual inscrease or decrease in the rate as it differs every year. As the bars differ every season, it might be possible to come to the conculsion that there has not been an immense change in physicality.*
